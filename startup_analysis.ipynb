```python
print("Missing Values in each column:")
print(df.isnull().sum())
```

```python
df = df.drop(columns=['permalink', 'homepage_url', 'name', 'state_code'])
print("Irrelevant columns dropped!")
```

```python
df['funding_total_usd'] = pd.to_numeric(df['funding_total_usd'].replace('-', np.nan), errors='coerce')

df['category_list'] = df['category_list'].fillna(df['category_list'].mode()[0])
df['country_code'] = df['country_code'].fillna(df['country_code'].mode()[0])
df['region'] = df['region'].fillna(df['region'].mode()[0])
df['city'] = df['city'].fillna(df['city'].mode()[0])

df['funding_total_usd'] = df['funding_total_usd'].fillna(df['funding_total_usd'].median())

df = df.dropna(subset=['founded_at', 'first_funding_at'])

print("All missing values handled!")
```

```python
print("Remaining missing values:")
print(df.isnull().sum())
print("\nNew shape of dataset:", df.shape)
```

```python
print("Number of duplicate rows:", df.duplicated().sum())
```

```python
df = df.drop_duplicates()
print("Duplicates removed!")
print("New shape of dataset:", df.shape)
```

```python
print("Remaining duplicates:", df.duplicated().sum())
```

```python
print(df.columns)
```

```python
df = df.rename(columns={
    'category_list': 'industry',
    'funding_total_usd': 'total_funding',
    'country_code': 'country',
    'state_code': 'state',
    'founded_at': 'founded_date',
    'first_funding_at': 'first_funding_date',
    'last_funding_at': 'last_funding_date',
})
print("Columns renamed successfully!")
print(df.columns)
```
